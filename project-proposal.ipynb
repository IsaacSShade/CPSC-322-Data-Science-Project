### Potential Impact of Results

1) Why are these results useful?
These results are useful because one could use this dataset to correctly identify an unknown mushroom. However, it is worth noting the severity of any false positives in our classification. Since a false positive could result in a poisonous mushroom being consumed, a false positive is much worse than a false negative, which is just an edible mushroom being tossed. With this in mind, we add emphasis on the False Positive Rate in our classifier analysis.
2) Who are stakeholders (users/groups that are impacted by the outcome of the project) interested in these results?
The stakeholders of this project would be any interested "shroomers" better known as mushroom hunters or foragers. Since they need to be well informed in their mushroom hunting, they could use this tool to classify any unknown mushrooms and determine whether they should be eaten or not.
