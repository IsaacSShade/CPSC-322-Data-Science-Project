### Implementation / Technical Merit

1) Anticipated challenges in pre-processing and/or classification (e.g., missing values, class distribution, etc.)


Loud cover and Daily summary are two classifiers that we could potentially use, but it appears to be text data that isn't easy to classify and use the classifiers on. There also appears to be flawed data with pressure (as some values say zero when they shouldn't say zero) so some cleaning may be necessary to clean the set before classification.

Another issue is that we cannot assume these variables to be independent, which is a critical component of Naive Bayes classifier. Since pressure and temperature or pressure and humidity may tend to move together, it is not an assumption we can make when trying to predict weather. This also means some classes may fall into distinct clusters, so we need to be particular in the classifiers that we use.

2) If the number of attributes is large, how will you explore feature selection techniques to pare down the attributes?


The full dataset has 12 columns with some dependent attributes, meaning we may have excessive attributes that don't actually provide new information (and may just overcomplicate our classifier). This implies we may only need to take a subset of the data to construct the most accurate classifier. Fortunately, we have classifer by classifier methods to deal with this issue, and we can evaluate by taking subsets of classifers and testing accuracy. We can test the two extremes (all features, or only a single feature) and compare results, and hopefully look for a perfect medium.
