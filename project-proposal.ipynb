### Implementation / Technical Merit

1) Anticipated challenges in pre-processing and/or classification (e.g., missing values, class distribution, etc.)


Something we noted in examination of the dataset is that the "bruises" column is entirely labeled as "mismatched" by Kaggle. However, we find every element in the column is labeled "t" or "f" while Kaggle wants them labeled as "True" or "False", which leads us to believe that the data is correct and Kaggle is reading it incorrectly. We will need to double check before we begin classification, but we believe that we may still use this feature without any issues on data mismatching. Otherwise the data is clean.

An assumption made in some classifiers (such as Naive Bayes) is that all features are independent of each other. This may not actually be the case, as some mushroom characteristics probably exist together with a fair amount of frequency (especially if only a few specific mushroom species are identified). This is something that we need to account for when we examine our classifiers, because this would also affect our class distribution and thus require us to do some stratification across folds.


2) If the number of attributes is large, how will you explore feature selection techniques to pare down the attributes?


The full dataset has 23 columns, meaning we may have excessive attributes that don't actually provide new information (and may just overcomplicate our classifier). Since the danger of a false positive is extreme, we need to be methodic about feature selection. Fortunately, we have classifer by classifier methods to deal with this issue, and we can evaluate by taking subsets of classifers and testing accuracy. We can test the two extremes (all features, or only a single feature) and compare results, and hopefully look for a perfect medium.

### Potential Impact of Results

1) __Why are these results useful?__  
These results are useful because one could use this dataset to correctly identify an unknown mushroom. However, it is worth noting the severity of any false positives in our classification. Since a false positive could result in a poisonous mushroom being consumed, a false positive is much worse than a false negative, which is just an edible mushroom being tossed. With this in mind, we add emphasis on the False Positive Rate in our classifier analysis.  
  
2) __Who are stakeholders (users/groups that are impacted by the outcome of the project) interested in these results?__    
The stakeholders of this project would be any interested "shroomers" better known as mushroom hunters or foragers. Since they need to be well informed in their mushroom hunting, they could use this tool to classify any unknown mushrooms and determine whether they should be eaten or not.
